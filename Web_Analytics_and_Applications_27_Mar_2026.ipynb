```python
# Phase 2: Web Personalization (User Profiling)

# Let’s begin with a simple question:
# “How does a website remember what you like?”
# The answer is User Profiling.

# User profiling is the process of creating a digital representation of a user based on their behavior, preferences, and interactions.
```

```python
# What is a Digital Profile?

# A digital profile contains:
# Interests (e.g., sports, movies, electronics)
# Behavior patterns
# Preferences
# Activity history

# On Amazon:
# If you:
# Search for laptops
# Compare prices
# Read reviews
# Your profile becomes:
# “User interested in electronics”
```

```python
# Problem Statement
# We have user browsing logs from a website.
# Our goal is to convert raw activity into a meaningful user profile
# that can be used for personalization and recommendation.

# Step 1:
import pandas as pd

data = pd.DataFrame({
    "user_id": [101,101,101,102,102,103,103,103,103],
    "timestamp": [
        "2026-03-20 10:00","2026-03-20 10:05","2026-03-20 10:10",
        "2026-03-20 11:00","2026-03-20 11:10",
        "2026-03-20 12:00","2026-03-20 12:05","2026-03-20 12:10","2026-03-20 12:20"
    ],
    "page_category": [
        "Electronics","Electronics","Accessories",
        "Fashion","Fashion",
        "Electronics","Mobiles","Mobiles","Accessories"
    ],
    "time_spent_sec": [60,120,40,200,150,80,100,120,60]
})

data["timestamp"] = pd.to_datetime(data["timestamp"])

data
```

```python
# Step 2: Feature Engineering (Industry Thinking)

# Feature 1: Total Time Spent per Category
category_time = data.groupby(["user_id","page_category"])["time_spent_sec"].sum().unstack(fill_value=0)

category_time
```

```python
# Feature 2: Number of Visits per Category

category_visits = data.groupby(["user_id","page_category"]).size().unstack(fill_value=0)

category_visits
```

```python
# Step 3: Combine Features into User Profile

user_profile = category_time.add_suffix("_time").join(
    category_visits.add_suffix("_visits")
)

user_profile
```

```python
# Step 4: Normalize to Get Interest Scores

normalized_profile = category_time.div(category_time.sum(axis=1), axis=0)

normalized_profile
```

```python
# Step 5: Identify Dominant Interest (Key Industry Insight)

user_profile["top_interest"] = normalized_profile.idxmax(axis=1)

user_profile
```

```python
# Step 1: Initialize Spark Session

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("UserProfilingPipeline") \
    .getOrCreate()

spark
```

```python
# Step 2:

from pyspark.sql import Row
from pyspark.sql.functions import col

data = [
    Row(user_id=101, timestamp="2026-03-20 10:00:00", category="Electronics", time_spent=60),
    Row(user_id=101, timestamp="2026-03-20 10:05:00", category="Electronics", time_spent=120),
    Row(user_id=101, timestamp="2026-03-20 10:10:00", category="Accessories", time_spent=40),

    Row(user_id=102, timestamp="2026-03-20 11:00:00", category="Fashion", time_spent=200),
    Row(user_id=102, timestamp="2026-03-20 11:10:00", category="Fashion", time_spent=150),

    Row(user_id=103, timestamp="2026-03-20 12:00:00", category="Electronics", time_spent=80),
    Row(user_id=103, timestamp="2026-03-20 12:05:00", category="Mobiles", time_spent=100),
    Row(user_id=103, timestamp="2026-03-20 12:10:00", category="Mobiles", time_spent=120),
    Row(user_id=103, timestamp="2026-03-20 12:20:00", category="Accessories", time_spent=60),
]

df = spark.createDataFrame(data)

df.show()
```

```python
# Step 3: Convert Timestamp to Proper Format

from pyspark.sql.functions import to_timestamp

df = df.withColumn("timestamp", to_timestamp("timestamp"))

df.printSchema()
```

```python
# Step 4: Feature Engineering

# Feature 1: Total Time Spent per Category
from pyspark.sql.functions import sum as spark_sum

category_time = df.groupBy("user_id", "category") \
    .agg(spark_sum("time_spent").alias("total_time"))

category_time.show()
```

```python
# Feature 2: Number of Visits

from pyspark.sql.functions import count

category_visits = df.groupBy("user_id", "category") \
    .agg(count("*").alias("visit_count"))

category_visits.show()
```

```python
# Step 5: Combine Features

user_features = category_time.join(
    category_visits,
    on=["user_id", "category"]
)

user_features.show()
```

```python
# Step 6: Normalize (Interest Score)

from pyspark.sql.window import Window
from pyspark.sql.functions import sum as spark_sum

window_spec = Window.partitionBy("user_id")

user_features = user_features.withColumn(
    "total_user_time",
    spark_sum("total_time").over(window_spec)
)

user_features = user_features.withColumn(
    "interest_score",
    col("total_time") / col("total_user_time")
)

user_features.show()
```
